```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                    "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                             pd.NaT]})
```

```python
df
```

```python
df.dropna() ##drops the Df rows containing na 
#if i write inplace = True in () it will modify the original df DataFrame
```

```python
df
```

```python
df.dropna(how = "all") ##it will not remove na unless all the elements in a particular row or column = na
```

```python
df.describe()
```

```python
df.info() ##information about the DataFrame
```
