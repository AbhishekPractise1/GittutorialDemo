## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline
##if you use this you need not use plot.show() in every line
```

```python
import numpy as np
```

```python
## Simple Examples

x=np.arange(0,10) ##creating an array starting from 0 to 9not including 10
y=np.arange(11,21)
```

```python
##plotting using matplotlib 

##plt scatter

plt.scatter(x,y,c='g')##used to scatter the value of x and y
## c = color "g " = green
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')

```

```python
y=x*x
```

```python
## plt plot

plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
#r* for red and stars
##place your pointer and click on the brackett and press shift + tab for description
plt.xlabel('X axis')##x axis name
plt.ylabel('Y axis') ##y axis name
plt.title('2d Diagram')##title of the plot
```

```python
## Creating Subplots
##subplots = whithin one canvas/block i can create multiple plots

plt.subplot(2,2,1)##plt.subplot(n of rows, n of columns,position)
plt.plot(x,y,'r--')#position 1 diagram
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')#position 2 diagram
plt.subplot(2,2,3)
plt.plot(x,y,'bo')#position 3 diagram
plt.subplot(2,2,4)
plt.plot(x,y,'go')#postion 4 diagram
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) ##sin is a fucntion tha will take the value of x and will find the sin of that value
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) ##formula
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1) ##2 rows 1 column
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) ##plotting a bar
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.grid(True)
plt.show()
##A histogram is a graphical display of data using bars of different heights. In a histogram, each bar groups numbers into ranges. Taller bars show that more data falls in that range.
##A histogram displays the shape and spread of continuous sample data.
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);  ##patch artist for filling box plot with colors
##ver = True...vertical representation
##helps in finding percentile
##study Statistics before plotting
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice
##explode == how much you want to seperate(how much far) a part from the pie chart

# Ploting 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)
##autopct enables you to display the percent value using Python string formatting.
#shadows shows shadow in piechart to improve the piechart representation 
plt.axis('equal')

plt.show()
```
